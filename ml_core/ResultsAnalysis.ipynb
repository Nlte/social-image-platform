```python
import pandas as pd
import numpy as np
from scipy import stats, integrate
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
sns.set(color_codes=True)
sns.set_context(rc={"figure.figsize": (20, 10)})
```

## Comparing benchmark and refined models

```python
df = pd.read_csv("results.csv")

benchmark = df[df["model"] == "benchmark"]
refined = df[df["model"] == "1hidden-1500-Adam5e-4"]

# label AUC
benchmark_label = benchmark.drop(["mean_auc", "model"], axis=1)
refined_label = refined.drop(["mean_auc", "model"], axis=1)
df_label = pd.concat([benchmark_label, refined_label]).T
df_label = df_label.rename(columns={0:"benchmark", 8:"refined"})

df_label.plot.bar()
```
